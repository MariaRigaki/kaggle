```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('training.csv', index_col='EventId')
df_test = pd.read_csv('test.csv', index_col='EventId')

labels = df['Label']
weights = df['Weight']
df_train = df.drop(['Label', 'Weight'], axis=1)
del df
```

```python
df2 = df_train.replace('-999.0', np.nan)
```

```python
df2.head()
```

```python
dft = df_test.replace('-999.0', np.nan)
np.shape(dft.dropna(axis=1))
```

```python
X = df2.fillna(df2.mean())
X.head()
del df2
```

```python
X_t = dft.fillna(dft.mean())
X_t.head()
del dft
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
X_scaled = scaler.transform(X)
X_t_scaled = scaler.transform(X_t)
```

```python
from sklearn.decomposition import RandomizedPCA
pca = RandomizedPCA(whiten=True)
pca.fit(X_scaled)
print(pca.explained_variance_ratio_) 
```

```python
from sklearn.decomposition import RandomizedPCA
pca = RandomizedPCA(n_components=15, whiten=True)
pca.fit(X)
```

```python
XX = pca.transform(X)
```

```python
X_t = pca.transform(X_t)
```

```python
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
clf = GradientBoostingClassifier(n_estimators=500, learning_rate=0.1, max_depth=6)
clf.fit(X, labels)
```

```python
import cPickle as pickle
pickle.dump(clf, open("clf_full2.pcl", "wb"))
```

```python
result = clf.predict(X_t)
predict_probs = clf.predict_proba(X_t)
probabilities = [max(l) for l in predict_probs]

sorted_probs = np.argsort(probabilities)

rank_order = list(sorted_probs)
for val,idx in zip(range(1, len(sorted_probs) + 1), sorted_probs):
    rank_order[idx] = val
```

```python
res_df = pd.DataFrame({'Class': result},
                       index=df_test.index.values,
                       #index_label='EventId',
                       columns=['RankOrder', 'Class'])
#res_df['EventId'] = df_train.index.values

res_df['RankOrder'] = rank_order

res_df.to_csv('result.csv')
```
